```python
#First objective is to find a specific ship and follow it through the dataset
MMSI = 240716000
chunks = pd.read_csv('../Data/aisdk-2025-01-19.csv', chunksize=chunk_size)
chunk = next(chunks)
cols = chunk.columns
print(cols)
cols_to_keep = [cols[0], cols[2], cols[3], cols[4]]

chunks = pd.read_csv('../Data/aisdk-2025-01-19.csv', chunksize=chunk_size, usecols=cols_to_keep)

ship_cols = [cols[0]] + cols_to_keep[1:] #I wanted to exclude MMSI but I failed
ship = pd.DataFrame(columns=ship_cols)

file_name = f"../Data/Single_ships/{MMSI}.csv"
if not os.path.exists(file_name):
    for i, chunk in enumerate(chunks):
        ship = pd.concat([ship, chunk[chunk['MMSI'] == MMSI][ship_cols]], ignore_index=True)
    ship.to_csv(file_name, index=False)
else:
    print(f"File {file_name} already exists!")
```
