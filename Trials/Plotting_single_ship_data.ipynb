```python
import pandas as pd
import sys
sys.path.append('../')

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
#importing the data
MMSI = 240716000
ship = pd.read_csv(f'../data/single_ships/{MMSI}.csv')
```

```python
ship.head()
```

```python
plt.plot(ship['Longitude'], ship['Latitude'])
```

```python
fig, ax = plt.subplots(subplot_kw={'projection': ccrs.PlateCarree()})

ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.OCEAN)

ax.plot(ship['Longitude'], ship['Latitude'])
```
